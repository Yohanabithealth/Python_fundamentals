## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

# Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**

3. Jumlah Major Category mana yang memiliki total value yang paling tinggi?

```python
top3 = df[['major_category', 'value']]
```

```python
top3_major_categories = top3.groupby('major_category').sum()
```

```python
top3_major_categories
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(top3_major_categories.index, top3_major_categories['value'], color='skyblue')
plt.title('Total Value for Top 3 Major Categories')
plt.xlabel('Major Category')
plt.ylabel('Total Value')

# Rotasi label sumbu x
plt.xticks(rotation=45, ha='right')  # ha='right' agar label lebih rapi

# Menambahkan label nilai di atas setiap bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, round(yval, 2), ha='center', va='bottom')

plt.show()
```

(**Insight**)

Insight yang dapat diambil dari visualisasi ini adalah major category yang paling tinggi berada pada pada Theft and Handing dengan total value sebesar 2661861

## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

**Question:**

5. Daerah manakah yang memiliki nilai box plot tertinggi?

```python
plt.figure(figsize=(10, 6))
plt.boxplot([top3_major_categories['value']], labels=['Top 3 Major Categories'], showfliers=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title('Box Plot of Total Values for Top 3 Major Categories')
plt.ylabel('Total Value')

plt.show()
```

**Insight:**

Dari visualisasi Box Plot dapat dilihat dari 3 top major categories menunjukkan nilai minimal adalah 0 dengan nilai Q1 sekitar 1, nilai q2 (median) sekitar 5 dan nilai Q3 adalah 7. Dan dari box plot dapat dilihat nilai max yaitu sekitar 15

## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

```python
import folium
import pandas as pd
from geopy.geocoders import Nominatim

# Data contoh
data = {
    'borough': ['Borough1', 'Borough2', 'Borough3'],
    'value': [100, 150, 200],  # Nilai total top 3 boroughs
}

# Membuat DataFrame
df_top3 = pd.DataFrame(data)

# Inisialisasi geolocator dari geopy
geolocator = Nominatim(user_agent="my_app")

# Menambahkan kolom 'location' yang berisi objek Location dari geopy
df_top3['location'] = df_top3['borough'].apply(lambda x: geolocator.geocode(x))

# Membuat peta dunia
world_map = folium.Map(location=[51.5074, -0.1278], zoom_start=11)  # Lokasi awal di London sebagai contoh

# Menambahkan penanda untuk setiap lokasi
for index, row in df_top3.iterrows():
    if row['location']:
        folium.Marker([row['location'].latitude, row['location'].longitude], 
                      popup=f"{row['borough']} - Value: {row['value']}").add_to(world_map)

# Menampilkan peta
world_map
```
